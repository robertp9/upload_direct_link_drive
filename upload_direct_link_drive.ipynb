## **1. SELECCIÓN DE DIRECTORIO**
Seleccione el directorio o carpeta, donde desea guardar los archivos.

```python
directorio_destino = 'TEST'
```

## **2. DAR PERMISOS DE GOOGLE DRIVE**
Para poder guardar sus archivos en el drive es necesario tener permisos del mismo.

```python
import urllib.request
import os
from google.colab import drive
drive.mount('/content/drive')

os.chdir(f'/content/drive/MyDrive/{directorio_destino}')
```

## **3. PARAMETROS DE LOS ARCHIVOS**
De algún nombre del archivo, para ser guardado

```python
nombre_archivo = "modelo_2011.pdf"
```

Base, es el prefijo numerico que se autoincrementa por cada descarga

```python
base = 0
```

## **4. URL A DESCARGAR**
Copie en la lista, la url(s) que desee descarga.

```python
urls = [
        "https://dialnet.unirioja.es/descarga/articulo/7554370.pdf",
        "https://dialnet.unirioja.es/descarga/articulo/5074389.pdf", 
       ]
```

## **5. INICIAR LA DESCARGA**
Este proceso copia de la url, al drive. Este proceso puede tardar cierto tiempo dependiendo del tamaño del archivo.

*Nota: Hay url que estan realacionas a token basados en tiempo o  tener direcciones de IP, donde se origino el enlace y al momento de descargar puede que no realice la descarga*

```python
for xurl in urls:
  urllib.request.urlretrieve(xurl,f"{base}_{nombre_archivo}")
  base = base + 1
```

## **6. COMPRIMIR ARCHIVOS**
Caso opcional, pero puede ser util, si desea guardar todos los archivos en comprimido tipo zip

```python
from zipfile import ZipFile
```

```python
data = os.listdir('./')

myzip = ZipFile('ZP-'+directorio_destino+'.zip', 'w')
for file in data:
    myzip.write(file)

myzip.close()
```
